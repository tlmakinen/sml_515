# problem 3

```python
def energy(total_drivers, drivers_on_path1):
    
    drivers_on_path2 = total_drivers - drivers_on_path1
    
    return (np.sum(np.arange(1, drivers_on_path1)) / 100)*2 + drivers_on_path2 * 5
```

```python
x = np.arange(1,301)
```

```python
y = [energy(total_drivers=300, drivers_on_path1=x) for x in x]
```

```python
import matplotlib.pyplot as plt
x = np.arange(0, 300)
plt.plot(x, y)
i1,i2 = np.where(y == min(y))[0]
plt.scatter(x[i1], y[i1], marker = '+',c='r', s=100, label='local PE min, $x$={}'.format(i1))
plt.scatter(x[i2], y[i2], marker = '+',c='r', s=100, label='local PE min, $x$= {}'.format(i2))
plt.legend(loc='best')
plt.xlabel('$x$ (Drivers on route ACDB)', fontsize=15)
plt.ylabel('Potential Energy of Traffic Pattern', fontsize=15)
```

# problem 4

define total energy function for original route + best-response route

```python
def energy2(total_drivers, drivers_on_path1):
    
    drivers_on_path2 = total_drivers - drivers_on_path1
    
    return np.sum(np.arange(1, drivers_on_path1)) + 5*(drivers_on_path1) + np.sum(np.arange(1,drivers_on_path2))*2
```

```python
x = np.arange(1, 9)
y = [energy2(total_drivers=9, drivers_on_path1=x) for x in x]
```

```python
x = np.arange(1, 9)
plt.plot(x, y)
i1 = np.where(y == min(y))[0][0]
plt.scatter(x[i1], y[i1], marker = '+',c='r', s=100, label='local PE min, $x$={}'.format(x[i1]))

plt.legend(loc='best')
plt.xlabel('$x$ (Drivers on route ADB)', fontsize=15)
plt.ylabel('Potential Energy of Traffic Pattern', fontsize=15)
```

## Problem 1: Design the Optimal Peak Detection System (3pts)

A common problem in the analysis event data is the precise localization of a peak with a known form, say a Gaussian. We generally have two options. We can either perform direct density estimation given the known parametric form, which is equivalent to a GMM with $K=1$. Or we can form a histogram, which turns event locations into counts per bin, and then fit a Gaussian to the pairs $(x_k, N_k)$ of (mean) bin location and counts. For image analysis, where the peak could be e.g. a blip on a radar monitor or an unresolved person on a satellite photo, we have don't have that choice because the incoming photons are automatically binned into pixels by our cameras.

### Step 1 (2pts)

Compare the uncertainty of the center estimation in these two cases as a function of the total event number $N$. Specifically, assume a standard normal distribution (i.e. $\sigma=1$) for the generating process. Draw $N$ events. Determine the center $\tilde\mu$ of the event distribution.

Then, bin the samples with a bin width $\Delta=1$. That gives you a set of bin centers and counts $\mathcal{D}=\lbrace (x_1,N_1), \dots, (x_K,N_K)\rbrace$. The likelihood for each bin is Poissonian, with a mean rate $\lambda_k$ that follows a Gaussian parametric form. But the Gaussian shape also needs to be integrated in the bins:

$$
\lambda_k = \frac{N}{\sqrt{2\pi}\sigma} \int_{x_k - \Delta/2}^{x_k + \Delta/2} \exp\left(-\frac{1}{2\sigma^2}(x - \mu)^2\right) dx = \frac{N}{2} \left[\mathrm{erf}\left(\frac{\Delta/2 - (x_k - \mu)}{\sqrt{2} \sigma}\right) + \mathrm{erf}\left(\frac{x_k - \mu + \Delta/2}{\sqrt{2} \sigma}\right)\right]
$$

for some unknown $\mu$. $\mathrm{erf}$ stands for the [Error function](https://en.wikipedia.org/wiki/Error_function). Use `scipy.optimize` to find the MLE $\tilde\mu_b$. 

Repeat the process multiple times, plot $p(\tilde\mu)$ and $p(\tilde\mu_b)$, and compute their variances. Repeat with different $N$.

## Step 1)

Begin with the construction of a direct estimator $\hat{\mu}$ of the center $\mu$ of the samples. Let $(X_i)_{i\in \{1,...,N\}}$ be iid random variables with 
$$ X_i \sim \mathcal{N}(\mu, \sigma^2),$$
with parameter $\mu$ and $\sigma$. We assume a known variance, so that we only have to estimate $\mu$. The MLE estimator $\hat{\mu}$ of $\mu$ is given by
$$ \hat{\mu} = \frac{1}{N}\sum_{i=1}^{N}X_i.$$
We therefore just have to calculate the sample average. 

Next, we consider a grid $(x_i)_{i\in\{1,...,K\}}$ with $x_1 \leq ... \leq x_K$ and set for $k\in\{1,...,K\}$
$$N_k = \sum_{i=1}^{N} \mathbb{1}_{\big\{x_k - 0.5\Delta \leq X_i < x_k + 0.5\Delta\big\}}.$$

Then, we know that 
$$ N_k \sim \text{Poisson}(\lambda_k),$$
with 
$$ \lambda_ k = \frac{N}{2} \left[\mathrm{erf}\left(\frac{\Delta/2 - (x_k - \mu)}{\sqrt{2} \sigma}\right) + \mathrm{erf}\left(\frac{x_k - \mu + \Delta/2}{\sqrt{2} \sigma}\right)\right].$$

Therefore, the maximum likelihood function $L_k(\mu)$ corresponding to the $k$-th bin is given by
$$ \log L_k(\mu) = N_k \log(\lambda_k) - \lambda_k.$$
We obtain the maximum likelihood estimator $\hat{\mu}_k$ corresponding to the $k$-th bin by
$$ \hat{\mu}_k = \text{arg}\max_{\mu} L_k(\mu).$$

We define the final estimator $\hat{\mu}_b$ of $\mu$ as
$$ \hat{\mu}_b = \frac{1}{K}\sum_{k=1}^{K}\hat{\mu}_k.$$


We implement the above for $S \in \mathbb{N}$ simulation to obtain a sample distribution of $\hat{\mu}$ and $\hat{\mu}_b$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
from scipy.stats import poisson, norm
from scipy.optimize import minimize


def sim(N = 100, K = 500, dt = 1.0):
    #number of samples
    #N = 100
    
    #number of simulations
    #K = 200
    
    X = np.random.normal(loc = 0, scale = 1, size = N*K).reshape((N,K))
    
    ##########################
    #Direct center
    ##########################
    
    mu = np.mean(X,axis = 0)
    #plt.hist(mu, 5, density = True)
    
    
    ##########################
    #Grid
    ##########################
    
    #dt = 1.0
    x = np.arange(np.min(X), np.max(X)+dt, dt)
    N_v = np.zeros((len(x), K))
    #x_grid = np.linspace(np.min(X), np.max(X)+ dt)
    
    for k in range(K):
        N_v[:,k] = np.vectorize(lambda x: np.sum( np.logical_and(X[:,k] >= x-0.5*dt,X[:,k] <x+0.5*dt)))(x)
        
    def loglikelihood(x_k, N_k, mu):
        
        lmbd_k = N/2*(special.erf((0.5*dt-x_k+mu)/np.sqrt(2))+special.erf((x_k-mu+0.5*dt)/np.sqrt(2)))
        return( N_k *np.log(lmbd_k) - lmbd_k )
           
    mu_b = np.zeros((len(x), K))
    for k in range(len(x)):
        for j in range(K):    
            mu_b[k,j] = minimize(lambda mu: -loglikelihood(x[k],N_v[k,j],mu), 0.0).x[0]
    mu_b = np.mean(mu_b, axis = 0)
    return mu, mu_b, X
```

We now plot sample distributions for different choices of $N$. We choose $N\in\{50,100,200,400,800\}$, $S = 500$ and $\Delta = 1$. 

```python
###############################################################################
#Plotting
###############################################################################
Ns = [50,100,200,400,800]
stats = np.zeros((len(Ns), 4))

for i,N in enumerate(Ns):
    mu, mu_b, X = sim(N = N)
    fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True)
    axes[0].hist(mu, density = True)
    axes[0].title.set_text("Histogram for mu, N = "+str(N)+", S= "+str(500))
    axes[1].hist(mu_b, density = True)
    axes[1].title.set_text("Histogram for mu_b, N = "+str(N)+", S= "+str(500))
    plt.show()
    stats[i, 0] = np.mean(mu)
    stats[i, 1] = np.mean(mu_b)
    stats[i, 2] = np.var(mu)
    stats[i, 3] = np.var(mu_b)
    
plt.plot(Ns, stats[:,2])
plt.plot(Ns, stats[:,3])
plt.title("Sample variance of estimators vs. N, S= "+str(500))
plt.ylabel("Sample Variance")
plt.xlabel("Number of events N")
plt.legend(("mu", "mu_b"))
plt.show()
```

We observe the following:

* The sample distribution of $\hat{\mu}$ folllows a normal distribution, which is something we mathematically expect. 
* The sample distribution of $\hat{\mu}_b$ looks somewhat normal. However, there are some spikes and non-normalities observable.
* The sample variance of $\hat{\mu}_b$ is larger than the sample variance of $\hat{\mu}$. This is something we expect given that by binning data, we give up some information.
